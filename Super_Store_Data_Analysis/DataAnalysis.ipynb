```python
import chardet
import pandas as pd

# Detect encoding
with open("C:\\Users\\saipr\\Downloads\\superstore_dataset2011-2015.csv", 'rb') as file:
    result = chardet.detect(file.read())
    encoding = result['encoding']

# Read the CSV file with detected encoding
df = pd.read_csv("C:\\Users\\saipr\\Downloads\\superstore_dataset2011-2015.csv", encoding=encoding)
```

```python
df=df.drop(['Row ID','Order ID','Customer ID','Postal Code'],axis=1)
```

```python
cat=df.groupby('Category',as_index=False)['Profit'].mean()
```

```python
import matplotlib as pyplot
cat.plot(kind='bar')
plt.title("Total profit by category")
plt.xlabel("category")
plt.ylabel("Total Profit")
plt.show()
```

```python
reg_sales=df.groupby('Region',as_index=False)['Profit'].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(x='Region',y='Profit',data=reg_sales)
```

```python
#Sales are higher during certain months of the year
df['Order Month']=pd.DatetimeIndex(df['Order Date']).month
```

```python
month_sales=df.groupby('Order Month')['Sales'].sum()
month_sales.plot()
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Orders with same day shipping have the lowest rate of returned products
df['Order day']=pd.DatetimeIndex(df['Order Date']).day
```

```python
daily_sales=df.groupby('Order day')['Sales'].sum()
daily_sales.plot()
plt.title("Total Sales by Day")
plt.xlabel("Day")
plt.ylabel("Total Sales")
plt.show()
```

```python
##Orders with same-day shipping have the lowest rate of returned products


# Calculate total orders by shipping mode
total_orders_by_shipping_mode = df.groupby('Ship Mode').size()

# Calculate returned orders by shipping mode (with negative profit)
returned_orders_by_shipping_mode = df[df['Profit'] < 0].groupby('Ship Mode').size()

# Ensure all shipping modes are included in the returned_orders_by_shipping_mode
returned_orders_by_shipping_mode = returned_orders_by_shipping_mode.reindex(total_orders_by_shipping_mode.index, fill_value=0)

# Calculate the percentage of returned orders by shipping mode
returned_per_by_shipping_mode = (returned_orders_by_shipping_mode / total_orders_by_shipping_mode) * 100

# Print the result
print(returned_per_by_shipping_mode)

# Plot the result
returned_per_by_shipping_mode.plot(kind="bar")
```

```python
#The company's profit is more on weekdays than weekends
import datetime
df['Order Day']=pd.DatetimeIndex(df['Order Date']).day_name()
day_profit=df.groupby('Order Day')['Profit'].sum()
day_profit.plot(kind="bar")
plt.title("Total Profit by the Day of week")
plt.xlabel("Day")
plt.ylabel("Total Sales")
plt.show()
```
