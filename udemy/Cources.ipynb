```python
#check Null Values
df.isnull().sum()
```

```python
import seaborn as sns
sns.heatmap(df.isnull())
```

```python
#Check for duplicated data and drop
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
#Find Out Number of Courses Per Subject
df.head(2)
df['subject'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,5))
sns.countplot(x='subject',data=df,palette=['green', 'yellow', 'pink', 'brown'])
plt.xlabel("Subjects",fontsize=13)
plt.ylabel("Number of Courses Per Subject",fontsize=13)
plt.xticks(rotation=60)
plt.show()
```

```python
#For which levels udemy providing the courses
df.head()
df['level'].value_counts()
```

```python
import matplotlib.pyplot as plt
sns.countplot(x='level',data=df,palette=['green', 'yellow', 'pink', 'brown'])
plt.xlabel("level",fontsize=13)
plt.ylabel("Number of Courses Per level",fontsize=13)
plt.xticks(rotation=60)
plt.show()
```

```python
#Display the count of paid and Free Courses
df.head()
df['is_paid'].value_counts()
```

```python
import matplotlib.pyplot as plt
sns.countplot(x='is_paid',data=df,palette=['green', 'yellow', 'pink', 'brown'])
plt.xlabel("Paid or Not",fontsize=13)
plt.ylabel("Number of paid and non paid courses",fontsize=13)
plt.xticks(rotation=60)
plt.show()
```

```python
#Which Free Course Has More Lectures(Free or Paid)
df.head()
df[df[df['is_paid']==False]['num_lectures'].max()==df['num_lectures']]
```

```python
#Which Cources have higher number of subscribers free or paid?
df.groupby('is_paid',as_index=False)['num_subscribers'].sum()
```

```python
sns.barplot(x='is_paid',y='num_subscribers',data=df)
```

```python
#Which Level Has The Highest Number of Subscribers?
df.head()
df.groupby('level',as_index=False)['num_subscribers'].max()
```

```python
sns.barplot(x='level',y='num_subscribers',data=df)
plt.xticks(rotation=45)
```

```python
#Find the most popular Course Title
df[df['num_subscribers'].max()==df['num_subscribers']]['course_title']
```

```python
df.groupby('course_title')['num_subscribers'].sum().sort_values(ascending=False).head(10)
```

```python
top_10=df.sort_values(by="num_subscribers",ascending=False).head(10)
```

```python
sns.barplot(x='num_subscribers',y='course_title',data=top_10)
```
